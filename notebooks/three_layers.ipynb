Bag-of-words
========

Both SFrames and SArrays expose functionality that can be very useful for manipulating text data. For example, one common preprocessing task for text data is to transform it into "bag-of-words" format: each document is represented by a map where the words are keys and the values are the number of occurrences. So a document containing the text "hello bloomberg hello" would be represented by a dict type element containing the value {"hello": 2, "bloomberg":1}. This transformation can be accomplished with the following code.

TF-IDF
=====

Another useful representation for text data is called TF-IDF (term frequency - inverse document frequency). This is a modification of the bag-of-words format where the counts are transformed into scores: words that are common across the document corpus are given low scores, and rare words occurring often in a document are given high scores.

TF-IDF(word,document)=N(word,document)âlog(1/âdN(word,d)))

where N(w, d) is the number of times word w occurs in document d. This transformation can be done to an SArray of dict type containing documents in bow-of-words format using tf_idf.

```python
#train_data, test_data = same_data.random_split(0.8, seed = 1234)

#level = 'Communications'


def training_classifiers(level, train_data, test_data):
    #Cascade approach
    #filter the training dataset by Coomunications at first level

    results_dic = {}
    #train_data, test_data = same_data.random_split(0.8)
    subset_train_data = train_data.filter_by(level, 'LVL1', exclude = False)
    #subset_communications_train_data.show()


    #train model using trigrams and bag of words features after removing stopwords
    model_level_one = gl.logistic_classifier.create(train_data, target='LVL1', features = ['trigrams', 'stop_bow'], class_weights = 'auto')
    predictions_level_one = model_level_one.predict(test_data)
    results_level_one = model_level_one.evaluate(test_data)
    
    

    #And first level predictions to test data and filter by predicted Communications
    test_data['predicted_level_one'] = predictions_level_one
    test_data_predicted_level_one = test_data.filter_by(level, 'predicted_level_one', exclude = False)
    test_data_predicted_level_one = test_data_predicted_level_one.filter_by(level, 'LVL1', exclude = False)

    #And train at level 2
    model_level_two = gl.logistic_classifier.create(subset_train_data, target='LVL2', features = ['trigrams', 'stop_bow'], class_weights = 'auto')
    predictions_level_two = model_level_two.predict(test_data_predicted_level_one)
    results_level_two = model_level_two.evaluate(test_data_predicted_level_one)
    
    results_dic[str(level)] = [results_level_one, results_level_two]
    
    return results_dic
```
